```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

#显示所有行
pd.set_option('display.max_rows', None)
#显示所有列
pd.set_option('display.max_columns', None)
```

# Tag数据

```python
train_tag_path = '训练数据集/训练数据集_tag.csv'
test_tag_path = '评分数据集b/评分数据集_tag_b.csv'

train_tag_df = pd.read_csv(train_tag_path)
test_tag_df = pd.read_csv(test_tag_path)

print(train_tag_df.shape)
print(test_tag_df.shape)
```

```python
X_tag = pd.concat([train_tag_df, test_tag_df])
```

## 处理NAN

```python
def replace_col_val(df, name):
    df[name].replace("\N", np.nan, inplace=True)
    df[name] = pd.to_numeric(df[name])
```

```python
num_with_n = ['age', 'job_year', 'frs_agn_dt_cnt', 'fin_rsk_ases_grd_cd', 'confirm_rsk_ases_lvl_typ_cd',
             'cust_inv_rsk_endu_lvl_cd', 'l6mon_daim_aum_cd', 'tot_ast_lvl_cd', 'pot_ast_lvl_cd',
              'bk1_cur_year_mon_avg_agn_amt_cd', 'l12mon_buy_fin_mng_whl_tms', 'l12_mon_fnd_buy_whl_tms',
             'l12_mon_insu_buy_whl_tms', 'l12_mon_gld_buy_whl_tms', 'pl_crd_lmt_cd', 'ovd_30d_loan_tot_cnt',
             'his_lng_ovd_day', 'hld_crd_card_grd_cd', 'l1y_crd_card_csm_amt_dlm_cd', 'perm_crd_lmt_cd',
             'cur_debit_cnt', 'cur_credit_cnt', 'cur_debit_min_opn_dt_cnt', 'cur_credit_min_opn_dt_cnt',
             'cur_debit_crd_lvl']

categorys = ['gdr_cd', 'mrg_situ_cd','edu_deg_cd','acdm_deg_cd', 'deg_cd','ic_ind', 'fr_or_sh_ind', 
             'dnl_mbl_bnk_ind', 'dnl_bind_cmb_lif_ind', 'hav_car_grp_ind', 'hav_hou_grp_ind', 
             'l6mon_agn_ind', 'vld_rsk_ases_ind', 'loan_act_ind', 'crd_card_act_ind']
```

```python
X_numerica = X_tag[num_with_n]
X_category = X_tag[categorys]
```

```python
# replace '\N' with np.nan
for name in num_with_n:
    replace_col_val(X_numerica, name)

X_category = pd.get_dummies(X_category)
```

```python
X_train_test_tag = pd.concat([X_numerica, X_category], axis=1)
print(X_train_test_tag.shape)
X_train_test_tag.reset_index(drop=True, inplace=True)
X_train_test_tag.head()
```

## 切出id, flag, 用于merge

```python
id_df = X_tag[['id', 'flag']].reset_index(drop=True)
id_df.tail()
```

# 交易数据 

```python
train_trd_path = '训练数据集/训练数据集_trd.csv'
test_trd_path = '评分数据集b/评分数据集_trd_b.csv'

train_trd_df = pd.read_csv(train_trd_path)
test_trd_df = pd.read_csv(test_trd_path)

print(len(train_trd_df))
print(len(test_trd_df))
```

```python
X_trd = pd.concat([train_trd_df, test_trd_df])
X_trd.reset_index(drop=True, inplace=True)
X_trd.head()
```

## 总的cny_trx_amt 统计指标 

```python
X_trd_amt_all = X_trd.groupby(['id'])['cny_trx_amt'].describe()
X_trd_amt_all.reset_index(inplace=True)
X_trd_amt_all.head()
```

## 基于Dat_Flg1_Cd 交易方向的统计指标

```python
X_trd_flg1_b = X_trd[X_trd['Dat_Flg1_Cd'] == 'B']
X_trd_flg1_c = X_trd[X_trd['Dat_Flg1_Cd'] == 'C']
```

```python
X_trd_amt_flg1_b = X_trd_flg1_b.groupby(['id'])['cny_trx_amt'].describe()
X_trd_amt_flg1_b.reset_index(inplace=True)
X_trd_amt_flg1_b.head()
```

```python
X_trd_amt_flg1_c = X_trd_flg1_c.groupby(['id'])['cny_trx_amt'].describe()
X_trd_amt_flg1_c.reset_index(inplace=True)
X_trd_amt_flg1_c.head()
```

## 结合交易方向（Dat_Flg1_Cd）& 支付手段（Dat_Flg3_Cd）

```python
print(X_trd['Dat_Flg1_Cd'].value_counts())
print(X_trd['Dat_Flg3_Cd'].value_counts())
```

```python
# example
# X_tmp = X_trd.loc[(X_trd['Dat_Flg1_Cd'] == 'B') & (X_trd['Dat_Flg3_Cd'] == 'A')].sample(5)
X_tmp.groupby(['id'])['cny_trx_amt'].describe().reset_index()
```

```python
flg1_flg3 = []
for flg1 in 'BC':
    for flg3 in 'ABC':
        X_tmp = X_trd.loc[(X_trd['Dat_Flg1_Cd'] == flg1) & (X_trd['Dat_Flg3_Cd'] == flg3)]
        flg13_feature = X_tmp.groupby(['id'])['cny_trx_amt'].describe().reset_index()
        flg1_flg3.append(flg13_feature)
print(len(flg1_flg3))
```

## 划分时间窗口 

```python
X_trd['trx_tm'] = pd.to_datetime(X_trd['trx_tm'])
```

### 基于月份 

```python
X_trd['month'] = X_trd['trx_tm'].apply(lambda x: x.month)
X_month = X_trd.groupby(['id', 'month'])['cny_trx_amt'].describe().unstack(level=-1)
X_month.reset_index(inplace=True)
X_month.head(5)
```

### 基于是否是工作日 

```python
X_trd['weekday'] = X_trd['trx_tm'].apply(lambda x: 1 if x.weekday() < 5 else 0)
X_weekday = X_trd.groupby(['id', 'weekday'])['cny_trx_amt'].describe().unstack(level=-1)
X_weekday.reset_index(inplace=True)
X_weekday.head()
```

### 基于一天内的时段 

```python
X_trd['hour_span'] = X_trd['trx_tm'].apply(lambda x: x.hour // 6)
X_hourspan = X_trd.groupby(['id', 'hour_span'])['cny_trx_amt'].describe().unstack(level=-1)
X_hourspan.reset_index(inplace=True)
X_hourspan.head()
```

### 是否在00：00 ~ 00：30交易（可能参与秒杀） 

```python
X_trd['seckill'] = X_trd['trx_tm'].apply(lambda x: 1 if x.hour == 0 and (0<= x.minute <= 30) else 0)
X_seckill = X_trd.groupby(['id', 'seckill'])['cny_trx_amt'].describe().unstack(level=-1)
X_seckill.reset_index(inplace=True)
X_seckill.head()
```

### 汇总基于交易行为的特征 

```python
trd_features = [id_df] + [X_trd_amt_all, X_trd_amt_flg1_b, X_trd_amt_flg1_c] + flg1_flg3 + [X_month, X_weekday, X_hourspan, X_seckill]
print(len(trd_features))
```

```python
import functools


def merge_by_id(df_a, df_b):
     return pd.merge(df_a, df_b, how='left', on=['id'])
```

```python
X_trd_features = functools.reduce(merge_by_id, trd_features)
X_trd_features.drop(columns=['id','flag'], inplace=True)
X_trd_features.fillna(0, inplace=True)
X_trd_features.head()
```

### k-fold 平滑预测结果

```python
from sklearn.model_selection import StratifiedKFold

params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'auc',
    'num_leaves': 40,
    'max_depth':6,
    'min_child_samples': 20,
    'min_child_weight': 0.001,
    'learning_rate': 0.05,
    'is_unbalance': 'true',
    'num_boost_round': 1000,
    'lambda_l1': 0.0,
    'lambda_l2': 0.0,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq':2,
    'cat_smooth': 0.0,
}

K = 5
kfold = StratifiedKFold(n_splits=K, shuffle=True, random_state=511).split(X_train_full, y_train_full)
print(X_test_full.shape[0] == 4000)
X_test = X_test_full.values
y_pred = np.zeros(X_test.shape[0])
```

```python
for i, (train_idx, val_idx) in enumerate(kfold):
    print('Process Flod: ', i)
    
    X_train = X_train_full.iloc[train_idx, :]
    X_val = X_train_full.iloc[val_idx, :]
    
    y_train = y_train_full.iloc[train_idx]
    y_val = y_train_full.iloc[val_idx]
    
    lgb_train = lgb.Dataset(X_train.values, y_train.values)
    lgb_valid = lgb.Dataset(X_val.values, y_val.values, reference=lgb_train)
    
    lgb_clf = lgb.train(params, lgb_train, valid_sets=lgb_valid, early_stopping_rounds=50)
    
    y_pred += lgb_clf.predict(X_test, num_iteration=lgb_clf.best_iteration)
```

```python
y_pred /= K
p_pred = pd.Series(y_pred)
result = pd.concat([test_tag_df['id'], p_pred], axis=1)
np.savetxt('result_lgb_tag_time_b_v.txt', result.values, fmt=('%s\t%.6f'))
result.head()
```
